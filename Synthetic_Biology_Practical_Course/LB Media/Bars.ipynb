# Lycopene 'C' and 'D' 

```python
plt.rc('xtick', labelsize=25)    # fontsize of the tick labels
plt.rc('ytick', labelsize=25)    # fontsize of the tick labels
plt.rcParams["font.family"] = "serif"


# set width of bar
barWidth = 0.35
fig = plt.figure(figsize=(16,9))

# set height of bar
lycopeneB1 = [2.38*1000000*(4.3/10)/(150479),1.9240990990990987*1000000*(4.3/10)/(150479),1.8122137404580154*1000000*(4.3/10)/(150479), 1.640857142857143*1000000*(4.3/10)/(150479)]
lycopeneB2 =[1.8868852459016394*1000000*(4.3/10)/(150479), 3.553140096618358*1000000*(4.3/10)/(150479),5.7120512820512825*1000000*(4.3/10)/(150479),0]
labels = ["0M","10$\mu$","100$\mu$","1m"]

# Set position of bar on X axis
br1 = np.arange(len(lycopeneB1))
br2 = [x + barWidth for x in br1]

# Make the plot

plt.bar(br1, lycopeneB1, color='k', width= barWidth, edgecolor= 'k', label='Plasmids C')
plt.bar(br2, lycopeneB2, color='crimson', width= barWidth, edgecolor= 'k', label='Plasmids D')
plt.ylabel('Lycopene ($\mu$M)', family = 'serif',fontsize = 35)
plt.xlabel('IPTG (M)', family = 'serif',fontsize = 35)

plt.xticks([r + barWidth -0.175 for r in range(len(lycopeneB1))], ["0M","10$\mu$","100$\mu$","1m"])
plt.text(-0.8, 5.7120512820512825*1000000*(4.3/10)/(150479), '(a)', fontsize = 40, family = 'serif')
plt.title(r'1.5$m$M Ara', family = 'serif',fontsize = 40)
plt.legend(prop={'size':30})
plt.savefig(f"{images_dir}/BarDC1_5Ara.png", bbox_inches="tight", dpi=100)
```
