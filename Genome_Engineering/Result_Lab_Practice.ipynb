# **Log and linear regression**

```python
Log_data23 = data23[:17]
Log_data23["Log23_A"] = np.log(Log_data23['A'])
Log_data23["Log23_B"] = np.log(Log_data23['B'])
Log_data23["Log23_C"] = np.log(Log_data23['C'])
Log_data23["Log23_D"] = np.log(Log_data23['D'])
Log_data23["Log23_E"] = np.log(Log_data23['E'])

Log_data24 = data24[:17]
Log_data24["Log24_A"] = np.log(Log_data24['A'])
Log_data24["Log24_B"] = np.log(Log_data24['B'])
Log_data24["Log24_C"] = np.log(Log_data24['C'])
Log_data24["Log24_D"] = np.log(Log_data24['D'])
Log_data24["Log24_E"] = np.log(Log_data24['E'])

Log_dataWT = dataWT[:17]
Log_dataWT["LogWT_A"] = np.log(Log_dataWT['A'])
Log_dataWT["LogWT_B"] = np.log(Log_dataWT['B'])
Log_dataWT["LogWT_C"] = np.log(Log_dataWT['C'])
Log_dataWT["LogWT_D"] = np.log(Log_dataWT['D'])
Log_dataWT["LogWT_E"] = np.log(Log_dataWT['E'])
```

## **Log 2+3**


```python
x= np.arange(8, len(Log_data23)*10,10)

#Optimization
from scipy.optimize import curve_fit 

#Fitting function
def func(x, a, b):
    return a*x+b
 
# Initial guess for the parameters
initialGuess = [1,1]    
 
#Perform the curve-fit
popt1, pcov1 = curve_fit(func, x/60,Log_data23["Log23_A"], initialGuess)
popt2, pcov2 = curve_fit(func, x/60,Log_data23["Log23_B"], initialGuess)
popt3, pcov3 = curve_fit(func, x/60,Log_data23["Log23_C"], initialGuess)
popt4, pcov4 = curve_fit(func, x/60,Log_data23["Log23_D"], initialGuess)
popt5, pcov5 = curve_fit(func, x/60,Log_data23["Log23_E"], initialGuess)
print('Slope 2+3 1: '+ str(popt1[0]) + '\n Y interset: '+ str(popt1[1]))
print('Slope 2+3 2: '+ str(popt2[0]) + '\n Y interset: '+ str(popt2[1]))
print('Slope 2+3 3: '+ str(popt3[0]) + '\n Y interset: '+ str(popt3[1]))
print('Slope 2+3 4: '+ str(popt4[0]) + '\n Y interset: '+ str(popt4[1]))
print('Slope 2+3 5: '+ str(popt5[0]) + '\n Y interset: '+ str(popt5[1]))

#Strain
S_23 = np.array([popt1[0],popt2[0],popt2[0],popt4[0]])

#x values for the fitted function
xFit = np.arange(0.0, 3.0, 0.01)
plt.figure(figsize=(12.5,6.6))

#Plot the fitted function
plt.plot(xFit, func(xFit, *popt1), color='indigo',lw=5.5)
plt.plot(xFit, func(xFit, *popt2), color='orange',lw=5.5)
plt.plot(xFit, func(xFit, *popt3), color='mediumspringgreen',lw=2.5)
plt.plot(xFit, func(xFit, *popt4), color='c',lw=2.5)
plt.plot(xFit, func(xFit, *popt5), color='k',lw=2.5)

plt.rc('xtick', labelsize=30)    # fontsize of the tick labels
plt.rc('ytick', labelsize=30)    # fontsize of the tick labels
plt.rcParams["font.family"] = "serif"
plt.plot(x/60, Log_data23["Log23_A"], ' ',color='k', marker='o', markersize=25,alpha=1,markerfacecolor="indigo")
plt.plot(x/60, Log_data23["Log23_B"], ' ',color='k', marker='o', markersize=25,markerfacecolor="orange")
plt.plot(x/60, Log_data23["Log23_C"], ' ',color='k', marker='o', markersize=25,markerfacecolor="mediumspringgreen")
plt.plot(x/60, Log_data23["Log23_D"], ' ',color='k', marker='o', markersize=25,markerfacecolor="c")
plt.plot(x/60, Log_data23["Log23_E"], ' ',color='k', marker='o', markersize=25,alpha=0.3, markerfacecolor="indigo")

plt.xlabel(r'Time (min)', family="serif",fontsize = 32)
plt.ylabel(r'$\log(\alpha_{growth})$', family="serif", fontsize = 32)
plt.title('Logarithm OD600', fontsize = 30, family = 'sarif')
plt.xlabel(r'Time (h)', family="serif",fontsize = 32)
plt.ylabel(r'log(OD 600)', family="serif", fontsize = 32)
#plt.legend(prop={"size":25, "family": "serif"},loc=(1.04,0.74))
plt.grid()
plt.savefig(f"{images_dir}/23LB_log.png", bbox_inches="tight", dpi=100)
```
