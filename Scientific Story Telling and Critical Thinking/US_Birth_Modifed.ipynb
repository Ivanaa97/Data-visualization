# **Weekly regularities**

```python
figura = births.pivot_table('Births', index='D-Week', columns='Year', aggfunc='mean')
```

```python
plt.rc('xtick', labelsize=11)    # fontsize of the tick labels
plt.rc('ytick', labelsize=11)    # fontsize of the tick labels
figura.plot(figsize=(11,8),style=':')
plt.gca().set_xticklabels(['', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('Mean births by day', fontsize=16)
plt.xlabel('Days of the week', fontsize=16)
```

```python
plt.rc('xtick', labelsize = 11)
plt.rc('xtick', labelsize = 11)

figura.plot(figsize=(11,8),style='o')
plt.gca().set_xticklabels(['', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('Mean births by day', fontsize=16)
plt.xlabel('Days of the week', fontsize=16)
```

```python
births2 = births[(births['Year']==2002) | (births['Year']==2004) | (births['Year']==2006) | (births['Year']==2008) | (births['Year']==2010)| (births['Year']==2012)| (births['Year']==2014)]
```

```python
births2 = births2.pivot_table('Births', index='D-Week', columns='Year', aggfunc='mean')
```

```python
plt.rc('xtick', labelsize = 14)
plt.rc('ytick', labelsize = 14)

births2.plot(figsize=(17,11),style=':', lw=3, legend='Year')
plt.gca().set_xticklabels(['', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('Mean births by day', fontsize=20,rotation=90)

plt.xlabel('Days of the week', fontsize=19)
plt.grid()
plt.legend(prop={"size":16})
```
