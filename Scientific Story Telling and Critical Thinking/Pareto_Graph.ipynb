```python
xm1=1
xm2=2
xm3=3
alpha1 = 3
alpha2 = 2
alpha3 = 1

xr1= np.arange(xm1,7,0.1)
ym1a1= alpha1*((xm1**(alpha1))/xr1**(alpha1 + 1))
ym1a2= alpha2*((xm1**(alpha2))/xr1**(alpha2 + 1))
ym1a3= alpha3*((xm1**(alpha3))/xr1**(alpha3 + 1))

xr2= np.arange(xm2,7,0.1)
ym2a1= alpha1*((xm2**(alpha1))/xr2**(alpha1 + 1))
ym2a2= alpha2*((xm2**(alpha2))/xr2**(alpha2 + 1))
ym2a3= alpha3*((xm2**(alpha3))/xr2**(alpha3 + 1))

xr3= np.arange(xm3,7,0.1)
ym3a1= alpha1*((xm3**(alpha1))/xr3**(alpha1 + 1))
ym3a2= alpha2*((xm3**(alpha2))/xr3**(alpha2 + 1))
ym3a3= alpha3*((xm3**(alpha3))/xr3**(alpha3 + 1))

plt.figure(figsize=(16,10))
plt.plot(xr1,ym1a1, color = 'teal',lw = 1.5,  label= r'$\alpha = 1$')
plt.plot(xr1,ym1a2, color = 'y',lw = 1.5, label= r'$\alpha = 2$')
plt.plot(xr1,ym1a3, color = 'purple',lw = 1.5,  label=r'$\alpha = 3$')

plt.plot(xr2,ym2a1, color = 'teal',lw = 1.5)
plt.plot(xr2,ym2a2, color = 'y',lw = 1.5)
plt.plot(xr2,ym2a3, color = 'purple',lw = 1.5)

plt.plot(xr3,ym3a1, color = 'teal',lw = 1.5)
plt.plot(xr3,ym3a2, color = 'y',lw = 1.5)
plt.plot(xr3,ym3a3, color = 'purple',lw = 1.5)
plt.plot(xm1)
plt.vlines(xm1, ymin=0, ymax=3,color='k',  ls=':',lw = 0.8)
plt.vlines(xm2, ymin=0, ymax=1.5,color='k', ls=':',lw = 0.8)
plt.vlines(xm3, ymin=0, ymax=1, color='k', ls=':',lw = 0.8)
plt.legend(loc='best', bbox_to_anchor=(1, 1),prop={"size":20})
plt.text(1, 3.01, r'$x_{m} = 1$', fontsize=17)
plt.text(2, 1.51, r'$x_{m} = 2$', fontsize=17)
plt.text(3, 1.01, r'$x_{m} = 3$', fontsize=17)
plt.xlabel('$x$', fontsize = '17')
plt.ylabel('$P(x)$', fontsize = '17')
```
